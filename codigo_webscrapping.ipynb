```python
from GoogleNews import GoogleNews
import pandas as pd

# Inicializamos GoogleNews en espa√±ol
googlenews = GoogleNews(lang='es', encode='utf-8')

# Establecer rango de fechas: todo 2023
googlenews.set_time_range('01/01/2023', '12/31/2023')

# Consultas espec√≠ficas
queries = [
    '"paro de docentes" Bogot√° "clases suspendidas"',
    '(paro OR "cese de actividades") AND (docentes OR maestros) AND Bogot√°',
    '"estudiantes sin clase" AND paro Bogot√°',
    '"suspensi√≥n de clases" AND docentes Bogot√°'
]

all_results = []

for q in queries:
    googlenews.search(q)
    results = googlenews.result()
    for r in results:
        all_results.append({
            'query': q,
            'title': r.get('title'),
            'media': r.get('media'),
            'date': r.get('date'),
            'link': r.get('link')
        })

# Guardar en DataFramedf = pd.DataFrame(all_results)print(df.head())

# Exportar a CSVdf.to_csv("paros_docentes_bogota_2023.csv", index=False, encoding="utf-8-sig")print("üìÅ Resultados guardados en paros_docentes_bogota_2023.csv")


import tempfile

# Guardar en DataFrame
df = pd.DataFrame(all_results)
print(df.head())

# Usar un directorio temporal para guardar el CSV
output_dir = tempfile.gettempdir()
output_path = f"{output_dir}/paros_docentes_bogota_2023.csv"
df.to_csv(output_path, index=False, encoding="utf-8-sig")
print(f"üìÅ Resultados guardados en {output_path}")
```

```python
# Lista de queries que quieres buscar
queries = [
    '("suspensi√≥n de clases" OR "cese de actividades" OR "interrupci√≥n acad√©mica") Bogot√° 2023',
    '("paro de maestros" OR "paro docente" OR "huelga de profesores" OR "protesta de docentes") Bogot√° 2023',
    '("interrupci√≥n de clases" OR "cancelaci√≥n de clases" OR "no hay clases") Bogot√° 2023',
    '("sindicato de maestros" OR "FECODE" OR "ADE Bogot√°") AND (paro OR huelga OR protesta) 2023',
    '("class suspension" OR "activity halt" OR "school shutdown") Bogot√° 2023',
    '("education strike" OR "teachers demonstration" OR "teachers march") Bogot√° 2023'
]

# Lista para almacenar todos los resultados
all_results = []

# üîç Iterar sobre cada query
for query in queries:
    print(f"\nüîç Buscando: {query}")

    # ‚öôÔ∏è Par√°metros de la B√∫squeda
    params = {
        "engine": "google",
        "q": f"{query} after:2020-01-01 before:2020-12-31",
        "api_key": SerpAPI_tkn1,
        "num": 50
    }

    # üöÄ Ejecutar la B√∫squeda
    try:
        search = GoogleSearch(params)
        results = search.get_dict()
        print("‚úÖ B√∫squeda completada.")

        # üìä Procesar Resultados Org√°nicos
        items = results.get("organic_results", [])

        # Agregar la columna 'query' a cada resultado
        for item in items:
            item['query_origen'] = query
            all_results.append(item)

    except Exception as e:
        print(f"‚ùå Error en query '{query}': {e}")
        continue

# üìä Crear DataFrame con todos los resultados
if all_results:
    df = pd.DataFrame(all_results)

    # üíæ Seleccionar columnas de inter√©s (incluyendo query_origen)
    columnas_disponibles = ["title", "link", "snippet", "query_origen"]
    columnas_existentes = [col for col in columnas_disponibles if col in df.columns]
    df = df[columnas_existentes]

    print("\n--- ¬°Resultados Encontrados! ---\n")
    print(df)
    print(f"\nTotal de resultados: {len(df)}")

    # Guardar en CSV
    df.to_csv("test_serpapi_simple_resultados.csv", index=False, encoding="utf-8-sig")
    print("\n------------------------------")
    print("üìÅ Resultados guardados en test_serpapi_simple_resultados.csv")

    # Mostrar conteo por query
    print("\nüìä Resultados por query:")
    print(df['query_origen'].value_counts())
else:
    print("\n‚ö†Ô∏è No se encontraron resultados en ninguna b√∫squeda.")
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import quote_plus
import time
from datetime import datetime

def buscar_en_eltiempo(termino_busqueda="paro maestros", max_resultados=15):
    """Busca noticias espec√≠ficas en El Tiempo usando requests"""
    print(f"\nüîç Buscando en El Tiempo: {termino_busqueda}")

    datos_paros = []

    # Codificar el t√©rmino de b√∫squeda para URL
    termino_codificado = quote_plus(termino_busqueda)

    # URL de b√∫squeda de El Tiempo
    url_busqueda = f"https://www.eltiempo.com/buscar/{termino_codificado}"

    # Headers para simular un navegador real
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
        'Accept-Language': 'es-ES,es;q=0.9,en;q=0.8',
        'Accept-Encoding': 'gzip, deflate, br',
        'Connection': 'keep-alive',
        'Upgrade-Insecure-Requests': '1'
    }

    try:
        # Hacer la petici√≥n
        response = requests.get(url_busqueda, headers=headers, timeout=10)
        response.raise_for_status()

        # Parsear el HTML
        soup = BeautifulSoup(response.content, 'html.parser')

        # Buscar art√≠culos - El Tiempo usa diferentes selectores
        # Intentamos varios selectores comunes
        articulos = (
            soup.find_all('article', limit=max_resultados) or
            soup.find_all('div', class_=lambda x: x and 'article' in x.lower(), limit=max_resultados) or
            soup.find_all('div', class_=lambda x: x and 'card' in x.lower(), limit=max_resultados)
        )

        print(f"üì∞ Encontrados {len(articulos)} art√≠culos")

        for articulo in articulos[:max_resultados]:
            try:
                # Extraer t√≠tulo
                titulo_elem = (
                    articulo.find('h2') or
                    articulo.find('h3') or
                    articulo.find(class_=lambda x: x and 'title' in x.lower())
                )
                titulo = titulo_elem.get_text(strip=True) if titulo_elem else ""

                # Extraer descripci√≥n
                descripcion_elem = (
                    articulo.find('p') or
                    articulo.find(class_=lambda x: x and ('description' in x.lower() or 'summary' in x.lower()))
                )
                descripcion = descripcion_elem.get_text(strip=True) if descripcion_elem else ""

                # Extraer link
                link_elem = articulo.find('a', href=True)
                link = link_elem['href'] if link_elem else ""

                # Asegurar que el link sea absoluto
                if link and not link.startswith('http'):
                    link = f"https://www.eltiempo.com{link}"

                # Extraer fecha si est√° disponible
                fecha_elem = (
                    articulo.find('time') or
                    articulo.find(class_=lambda x: x and ('date' in x.lower() or 'fecha' in x.lower()))
                )
                fecha = fecha_elem.get_text(strip=True) if fecha_elem else 'N/A'

                # Validar que tengamos al menos t√≠tulo y que sea significativo
                if titulo and len(titulo) > 10:
                    datos_paros.append({
                        'titulo': titulo,
                        'fuente': 'El Tiempo',
                        'fecha_publicacion': fecha,
                        'descripcion': descripcion,
                        'url': link,
                        'periodo_busqueda': termino_busqueda,
                        'fecha_extraccion': datetime.now().strftime("%Y-%m-%d %H:%M:%S")
                    })

                    print(f"  ‚Ä¢ {titulo[:60]}...")

            except Exception as e:
                print(f"  ‚ö†Ô∏è Error procesando art√≠culo: {e}")
                continue

        print(f"\n‚úÖ Extracci√≥n completada: {len(datos_paros)} noticias v√°lidas")

    except requests.exceptions.RequestException as e:
        print(f"‚ùå Error de conexi√≥n: {e}")
    except Exception as e:
        print(f"‚ùå Error general: {e}")

    return datos_paros


# Uso del c√≥digo
if __name__ == "__main__":
    # Ejemplo de uso
    resultados = buscar_en_eltiempo("paro maestros", max_resultados=15)

    # Convertir a DataFrame si tienes resultados
    if resultados:
        import pandas as pd
        df = pd.DataFrame(resultados)
        print("\nüìä DataFrame creado:")
        print(df[['titulo', 'fecha_publicacion', 'fuente']])

        # Guardar en CSV
        df.to_csv("noticias_eltiempo.csv", index=False, encoding="utf-8-sig")
        print("\nüíæ Guardado en noticias_eltiempo.csv")
```

```python
def buscar_en_elespectador(termino_busqueda="paro maestros", max_resultados=15):
    """Busca noticias en El Espectador usando requests"""
    print(f"\nüîç Buscando en El Espectador: {termino_busqueda}")

    datos_paros = []

    # Codificar el t√©rmino de b√∫squeda para URL
    termino_codificado = quote_plus(termino_busqueda)

    # El Espectador tiene b√∫squeda directa
    url_busqueda = f"https://www.elespectador.com/buscar/?q={termino_codificado}"

    # Tambi√©n puedes usar tags espec√≠ficos si los conoces:
    # url_busqueda = "https://www.elespectador.com/tags/paro-de-maestros/"

    # Headers para simular un navegador real
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
        'Accept-Language': 'es-CO,es;q=0.9,en;q=0.8',
        'Accept-Encoding': 'gzip, deflate, br',
        'Connection': 'keep-alive',
        'Upgrade-Insecure-Requests': '1',
        'Referer': 'https://www.elespectador.com/'
    }

    try:
        # Hacer la petici√≥n con sesi√≥n para mantener cookies
        session = requests.Session()
        response = session.get(url_busqueda, headers=headers, timeout=15)
        response.raise_for_status()

        # Parsear el HTML
        soup = BeautifulSoup(response.content, 'html.parser')

        # El Espectador usa diferentes estructuras - intentamos varios selectores
        articulos = (
            soup.find_all('article', limit=max_resultados) or
            soup.find_all('div', class_=lambda x: x and 'Card' in str(x), limit=max_resultados) or
            soup.find_all('div', class_=lambda x: x and 'article' in str(x).lower(), limit=max_resultados)
        )

        print(f"üì∞ Encontrados {len(articulos)} art√≠culos")

        for articulo in articulos[:max_resultados]:
            try:
                # Extraer t√≠tulo
                titulo_elem = (
                    articulo.find('h2') or
                    articulo.find('h3') or
                    articulo.find(class_=lambda x: x and 'Card-title' in str(x)) or
                    articulo.find(class_=lambda x: x and 'title' in str(x).lower())
                )
                titulo = titulo_elem.get_text(strip=True) if titulo_elem else ""

                # Extraer fecha
                fecha_elem = (
                    articulo.find('time') or
                    articulo.find(class_=lambda x: x and 'date' in str(x).lower()) or
                    articulo.find(class_=lambda x: x and 'Card-date' in str(x))
                )

                if fecha_elem:
                    # Intentar obtener el atributo datetime primero
                    fecha = fecha_elem.get('datetime', fecha_elem.get_text(strip=True))
                else:
                    fecha = "N/A"

                # Extraer descripci√≥n
                descripcion_elem = (
                    articulo.find('p') or
                    articulo.find(class_=lambda x: x and 'description' in str(x).lower()) or
                    articulo.find(class_=lambda x: x and 'summary' in str(x).lower())
                )
                descripcion = descripcion_elem.get_text(strip=True) if descripcion_elem else ""

                # Extraer link
                link_elem = articulo.find('a', href=True)
                link = link_elem['href'] if link_elem else ""

                # Asegurar que el link sea absoluto
                if link:
                    if not link.startswith('http'):
                        link = f"https://www.elespectador.com{link}"

                # Validar que tengamos t√≠tulo v√°lido
                if titulo and len(titulo) > 10:
                    datos_paros.append({
                        'titulo': titulo,
                        'fuente': 'El Espectador',
                        'fecha_publicacion': fecha,
                        'descripcion': descripcion,
                        'url': link,
                        'periodo_busqueda': termino_busqueda,
                        'fecha_extraccion': datetime.now().strftime("%Y-%m-%d %H:%M:%S")
                    })

                    print(f"  ‚Ä¢ {titulo[:60]}...")

            except Exception as e:
                print(f"  ‚ö†Ô∏è Error procesando art√≠culo: {e}")
                continue

        print(f"\n‚úÖ Extracci√≥n completada: {len(datos_paros)} noticias v√°lidas")

    except requests.exceptions.RequestException as e:
        print(f"‚ùå Error de conexi√≥n: {e}")
    except Exception as e:
        print(f"‚ùå Error general: {e}")

    return datos_paros


def buscar_en_elespectador_por_tag(tag_url, termino_busqueda="paro maestros", max_resultados=15):
    """Busca noticias en El Espectador usando una URL de tag espec√≠fica"""
    print(f"\nüîç Buscando en El Espectador por tag: {tag_url}")

    datos_paros = []

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
        'Accept-Language': 'es-CO,es;q=0.9,en;q=0.8',
        'Accept-Encoding': 'gzip, deflate, br',
        'Connection': 'keep-alive',
        'Referer': 'https://www.elespectador.com/'
    }

    try:
        session = requests.Session()
        response = session.get(tag_url, headers=headers, timeout=15)
        response.raise_for_status()

        soup = BeautifulSoup(response.content, 'html.parser')

        articulos = (
            soup.find_all('article', limit=max_resultados) or
            soup.find_all('div', class_=lambda x: x and 'Card' in str(x), limit=max_resultados)
        )

        print(f"üì∞ Encontrados {len(articulos)} art√≠culos")

        for articulo in articulos[:max_resultados]:
            try:
                titulo_elem = (
                    articulo.find('h2') or
                    articulo.find('h3') or
                    articulo.find(class_=lambda x: x and 'Card-title' in str(x))
                )
                titulo = titulo_elem.get_text(strip=True) if titulo_elem else ""

                fecha_elem = (
                    articulo.find('time') or
                    articulo.find(class_=lambda x: x and 'Card-date' in str(x))
                )
                fecha = fecha_elem.get('datetime', fecha_elem.get_text(strip=True)) if fecha_elem else "N/A"

                link_elem = articulo.find('a', href=True)
                link = link_elem['href'] if link_elem else ""

                if link and not link.startswith('http'):
                    link = f"https://www.elespectador.com{link}"

                if titulo and len(titulo) > 10:
                    datos_paros.append({
                        'titulo': titulo,
                        'fuente': 'El Espectador',
                        'fecha_publicacion': fecha,
                        'descripcion': '',
                        'url': link,
                        'periodo_busqueda': termino_busqueda,
                        'fecha_extraccion': datetime.now().strftime("%Y-%m-%d %H:%M:%S")
                    })

                    print(f"  ‚Ä¢ {titulo[:60]}...")

            except Exception as e:
                continue

        print(f"\n‚úÖ Extracci√≥n completada: {len(datos_paros)} noticias v√°lidas")

    except Exception as e:
        print(f"‚ùå Error: {e}")

    return datos_paros


# Uso del c√≥digo
if __name__ == "__main__":
    # Opci√≥n 1: Buscar por t√©rmino
    resultados1 = buscar_en_elespectador("paro maestros", max_resultados=15)

    # Opci√≥n 2: Buscar por tag espec√≠fico
    resultados2 = buscar_en_elespectador_por_tag(
        "https://www.elespectador.com/tags/paro-de-maestros/",
        termino_busqueda="paro maestros"
    )

    # Combinar resultados
    todos_resultados = resultados1 + resultados2

    if todos_resultados:
        import pandas as pd
        df = pd.DataFrame(todos_resultados)
        df = df.drop_duplicates(subset=['url'])  # Eliminar duplicados

        print("\nüìä DataFrame creado:")
        print(df[['titulo', 'fecha_publicacion', 'fuente']])

        # Guardar en CSV
        df.to_csv("noticias_elespectador.csv", index=False, encoding="utf-8-sig")
        print(f"\nüíæ Guardado {len(df)} noticias en noticias_elespectador.csv")
```
