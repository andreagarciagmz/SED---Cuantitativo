## Google News

```python
from GoogleNews import GoogleNews
import pandas as pd

# Inicializamos GoogleNews en español
googlenews = GoogleNews(lang='es', encode='utf-8')

# Establecer rango de fechas: todo 2023
googlenews.set_time_range('01/01/2023', '12/31/2023')

# Consultas específicas
queries = [
    '"paro de docentes" Bogotá "clases suspendidas"',
    '(paro OR "cese de actividades") AND (docentes OR maestros) AND Bogotá',
    '"estudiantes sin clase" AND paro Bogotá',
    '"suspensión de clases" AND docentes Bogotá'
]

all_results = []

for q in queries:
    googlenews.search(q)
    results = googlenews.result()
    for r in results:
        all_results.append({
            'query': q,
            'title': r.get('title'),
            'media': r.get('media'),
            'date': r.get('date'),
            'link': r.get('link')
        })

# Guardar en DataFramedf = pd.DataFrame(all_results)print(df.head())

# Exportar a CSVdf.to_csv("paros_docentes_bogota_2023.csv", index=False, encoding="utf-8-sig")print("📁 Resultados guardados en paros_docentes_bogota_2023.csv")


import tempfile

# Guardar en DataFrame
df = pd.DataFrame(all_results)
print(df.head())

# Usar un directorio temporal para guardar el CSV
output_dir = tempfile.gettempdir()
output_path = f"{output_dir}/paros_docentes_bogota_2023.csv"
df.to_csv(output_path, index=False, encoding="utf-8-sig")
print(f"📁 Resultados guardados en {output_path}")
```

```python
options = Options()
options.add_argument("--headless")  # No mostrar la ventana del navegador
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")
# Inicializar el navegador Chrome
driver = webdriver.Chrome(options=options)
# --- Definir la búsqueda ---
query = "paro docente 2023 Bogotá"
url = f"https://www.google.com/search?q={query.replace(' ', '+')}"

# Abrir Google y esperar a que cargue
driver.get(url)
time.sleep(2)  # Esperar 2 segundos para asegurar que cargue el contenido

# --- Extraer el HTML ---
soup = BeautifulSoup(driver.page_source, "html.parser")

# --- Buscar resultados ---
results = []
for item in soup.select('div.g'):
    title = item.select_one('h3')
    link = item.select_one('a')
    snippet = item.select_one('div.VwiC3b')

    if title and link and snippet:
        results.append({
            'title': title.text,
            'link': link['href'],
            'snippet': snippet.text
        })

# Cerrar el navegador
driver.quit()

# --- Crear DataFrame ---
df = pd.DataFrame(results)

# Mostrar primeras filas
print(df.head())

# Guardar CSV
df.to_csv("resultados_selenium_google.csv", index=False, encoding="utf-8-sig")
print("📁 Resultados guardados en resultados_selenium_google.csv")
```
